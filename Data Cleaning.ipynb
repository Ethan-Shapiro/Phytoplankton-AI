```python
cast["Date"] = cast["Date"].apply(lambda x: dt.datetime.strptime(x.replace("/", "-"), '%m-%d-%Y'))
cast["Date"]
```

```python
phyto.head()
```

```python
phyto.columns
```

```python
cols_keep = ['studyName', 'Bottle Number', 'Latitude (º)',
       'Longitude (º)', 'Line', 'Station', 'Diatom < 5um (µg/l)',
       'Diatom 5-10um (µg/l)', 'Diatom 10-20um (µg/l)',
       'Diatom 20-40um (µg/l)', 'Diatom > 40um (µg/l)', 'Total Diatom (µg/l)', "Depth (m)"]
```

```python
phyto = phyto[cols_keep]
```

```python
phyto.info()
```

```python
phyto = phyto.dropna(subset=['Depth (m)'], axis=0)
```

```python
phyto.info()
```

```python
phyto["Depth (m)"] = phyto["Depth (m)"].apply(int)
```

```python
phyto.info()
```

```python
phyto.columns = ['studyName', "Bottle Number", 'Latitude', 'Longitude', 'Line', 'Station', 'Diatom < 5um',
       'Diatom 5-10um', 'Diatom 10-20um', 'Diatom 20-40um', 'Diatom > 40um', 'Total Diatom', "Depth (m)"]
```

```python
df = df[df["Depth (m)"] <= 125]
```

```python
df["Depth (m)"].describe()
```

```python
phyto["Depth (m)"].describe()
```

```python
N = 100
df["Longitude"] = np.round(df.Longitude*N).astype(int) 
df["Latitude"] = np.round(df.Latitude*N).astype(int) 

phyto["Longitude"] = np.round(phyto.Longitude*N).astype(int) 
phyto["Latitude"] = np.round(phyto.Latitude*N).astype(int)
```

```python
final = pd.merge(df, phyto, how="inner", on=["studyName", "Bottle Number", "Depth (m)", "Longitude", "Latitude"])
```

```python
final
```

```python
final.isna().sum()
```

```python
FINAL = final.dropna(subset=['Diatom < 5um', 'Diatom 5-10um', 'Diatom 10-20um', 'Diatom 20-40um', 
                      'Diatom > 40um', 'Total Diatom', "SiO3uM"])
```

```python
FINAL.isna().sum()
```

```python
FINAL = FINAL.drop(["Line_x", "Station_x", "Line_y", "Station_y"], axis=1)
```

```python
FINAL.to_csv("phytoplankton_final_dataset.csv", index=False)
```
